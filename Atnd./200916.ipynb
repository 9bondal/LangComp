```python
def article(word:str):
    if word.startswith(("a","e","i","o","u")):
        return "an " + word
    else:
        return "a " + word

print(article("aword"))
```

```python
basket = ["bacon","egg","bacon","ham","cheese","egg", "bacon"]
freq = dict()

for item in basket:
    freq[item] = freq.get(item, 0) + 1

freq
```
